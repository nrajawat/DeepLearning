```python
# [9.6] Extending...Combined visualisation for above plot

# Calculate the correlation coefficients between each attribute and Medv
correlations = data.corr()['Medv']

# Sort the correlations in descending order
sorted_correlations = correlations.sort_values(ascending=False)

# Find variables positively correlated with Medv
positively_correlated_vars = sorted_correlations[sorted_correlations > 0]

# Find variables negatively correlated with Medv
negatively_correlated_vars = sorted_correlations[sorted_correlations < 0]

# Combine positively and negatively correlated variables
all_correlated_vars = pd.concat([positively_correlated_vars, negatively_correlated_vars])

# Plot the correlated variables
plt.figure(figsize=(10, 5))
plt.bar(all_correlated_vars.index, all_correlated_vars.values)
plt.xlabel('Attribute')
plt.ylabel('Correlation with Medv')
plt.title('Variables (Positively & Negatively) Correlated with Medv')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
# [9.7] 
# Draw scatterplots to show relationship between each attribute and Medv, respectively. [0.25]

# Load the dataset as done previously in data


# For Getting list of attributes
Attributes = data.columns[:-1]

# To Set up the subplot grid
num_plots = len(Attributes)
cols = 3  # Number of columns in subplot grid
rows = (num_plots - 1) // cols + 1  # Number of rows in the subplot grid

# Create the figure and axes for the subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, 20))

# Flatten the axes array
axes = axes.flatten()

# For scatterplots for each attribute
for i, attr in enumerate(Attributes):
    ax = axes[i]
    ax.scatter(data[attr], data['Medv'])
    ax.set_xlabel(attr)
    ax.set_ylabel('Medv')

# For Removing unused plots
for j in range(num_plots, len(axes)):
    fig.delaxes(axes[j])

# For subplots layout Adjust
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# [9.8]
# Explain how to use scatterplots to find attributes which are positively correlated, 
# negatively correlated, or independent of Medv, respectively [0.25]



# Load the dataset and extract all columns (previously done)


# Get the list of column names
columns = data.columns

# Create a grid for the scatter plots
num_plots = len(columns) - 1  # Excluding the target variable
num_cols = 2  # Number of columns in the grid
num_rows = (num_plots + num_cols - 1) // num_cols # Number of rows in the grid

# Create the figure and axes for the subplots
Fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 30))

# For easier iterationFlatten the axes array 
axes = axes.flatten()

# Iterate over each column except the target variable
for i, column in enumerate(columns[:-1]):
    ax = axes[i]  # For Getting the current subplot

    # Scatter plot
    ax.scatter(data[column], data['Medv'])
    ax.set_xlabel(column)
    ax.set_ylabel('Medv')

    # Calculate the correlation coefficient
    correlation = data[column].corr(data['Medv'])

    # Determine the correlation type
    if correlation > 0.5:
        correlation_type = 'Positively correlated'
        color = 'green'
    elif correlation < -0.5:
        correlation_type = 'Negatively correlated'
        color = 'red'
    else:
        correlation_type = 'Independent'
        color = 'blue'

    # Add correlation type to the plot
    ax.text(0.5, 0.9, correlation_type, transform=ax.transAxes,
            fontsize=20, color=color, ha='center')

# For Removing unused subplots
if num_plots < len(axes):
    for j in range(num_plots, len(axes)):
        fig.delaxes(axes[j])

# For spacing adjustment between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
